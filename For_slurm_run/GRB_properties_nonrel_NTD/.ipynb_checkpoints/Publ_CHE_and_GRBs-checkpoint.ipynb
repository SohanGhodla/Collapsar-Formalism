```python
def percent_pathway(directory, file):
    df = pd.read_csv(directory + file, delim_whitespace=True,header = None)

    T = []
    for w in np.array_split(df, 3): # 1 = BS, 2 = primary, 3= secondary
        temp = []
        for x in np.array_split(w, 7): # CHE type, 1 = tidal, 2 = accretion, 3 = no CHE, 4 ..5 ..6..
            temp1 = []
            for y in np.array_split(x, 2): # 1 = No BBH, 2 = BBH
                temp2 = []
                for z in np.array_split(y, 2): #0 = No L2OF
                    temp2.append(np.array_split(z,32)) # 32 allowed period 
                temp1.append(temp2)
            temp.append(temp1)
        T.append(temp)
        
        
    total_mergers = 0
    #To find the total number of BBH formed
    for i in range(3):
        for p in range(32): #over period
            for M in range(1,41): # over iMtot 
                total_mergers += (sum(T[0][i][1][0][p].iloc[0:51, M]) + sum(T[0][i][0][0][p].iloc[0:51, M]))
                
    total_tidal_CHE_both = 0
    total_tidal_CHE_pri = 0
    total_tidal_CHE_sec = 0
    total_accrt_CHE = 0
    No_CHE = 0
    total_accrt_CHE_tidally_spindown = 0
    for p in range(32): #over period
        for M in range(1,41): # over iMtot 
            total_tidal_CHE_both += (sum(T[0][3][1][0][p].iloc[0:51, M]) + sum(T[0][3][0][0][p].iloc[0:51, M]))
            total_tidal_CHE_pri += (sum(T[0][4][1][0][p].iloc[0:51, M]) + sum(T[0][4][0][0][p].iloc[0:51, M]))
            total_tidal_CHE_sec += (sum(T[0][5][1][0][p].iloc[0:51, M]) + sum(T[0][5][0][0][p].iloc[0:51, M]))
            total_accrt_CHE += (sum(T[0][1][1][0][p].iloc[0:51, M]) + sum(T[0][1][0][0][p].iloc[0:51, M]))
            total_accrt_CHE_tidally_spindown += (sum(T[0][6][1][0][p].iloc[0:51, M]) \
                                                             + sum(T[0][6][0][0][p].iloc[0:51, M]))
            No_CHE += (sum(T[0][2][1][0][p].iloc[0:51, M]) + sum(T[0][2][0][0][p].iloc[0:51, M]))
            
    
    return [total_mergers, total_tidal_CHE_both, total_tidal_CHE_pri, total_tidal_CHE_sec, \
            total_accrt_CHE, No_CHE, total_accrt_CHE_tidally_spindown]
```

```python
def Mass_vs_Z_pathway(directory, file):
    df = pd.read_csv(directory + file, delim_whitespace=True,header = None)

    T = []
    for w in np.array_split(df, 3): # 1 = BS, 2 = primary, 3= secondary
        temp = []
        for x in np.array_split(w, 7): # CHE type, 1 = tidal, 2 = accretion, 3 = no CHE, 4 ..5 ..6..
            temp1 = []
            for y in np.array_split(x, 2): # 1 = No BBH, 2 = BBH
                temp2 = []
                for z in np.array_split(y, 2): #0 = No L2OF
                    temp2.append(np.array_split(z,32)) # 32 allowed period 
                temp1.append(temp2)
            temp.append(temp1)
        T.append(temp)
            
    total_mergers =  np.array([0]*40)        
    #To find the total number of BBH formed
    for i in range(3):
        for p in range(32): #over period
            for M in range(1,41): # over iMtot 
                total_mergers[M - 1] += (sum(T[0][i][1][0][p].iloc[0:51, M]) + sum(T[0][i][0][0][p].iloc[0:51, M]))
    
    total_accrt_CHE = np.array([0]*40)        
    for p in range(32): #over period
        for M in range(1,41): # over M2_ZAMS   2 below becaue want secondarys mass
            total_accrt_CHE[M - 1] += (sum(T[2][1][1][0][p].iloc[0:51, M]) + sum(T[2][1][0][0][p].iloc[0:51, M]))
                   
    return [total_mergers, total_accrt_CHE,]
```

```python
def count_ionising_photons(directory, file, index, how_many):
    df = pd.read_csv(directory + file, delim_whitespace=True,header = None)

    T = []
    for w in np.array_split(df, 3):
        T.append(w)
            
    total_photons =  [0]*51#[ [0] for i in range(51)] # np.array([0]*51)        
    
    if how_many < 3: # if index = 3 then sum all.
        for j in range(51):
            total_photons[j] += T[index].iloc[j,0]
            
    elif how_many ==4: # if index = 3 then sum all.
        for i in range(2):
            for j in range(51):
                total_photons[j] += T[i].iloc[j,0]
    else:
        for i in range(3):
            for j in range(51):
                total_photons[j] += T[i].iloc[j,0]
            
    # Binning them
    # integrated photons hold the total number of photons till 100 Mys.
    integrated_photons = 0
    delaytime = []
    for itime in range(51):
        delaytime.append(6 + 0.1*itime)
        
        if itime == 0:
            del_t = 10**(6.05) #calculating the time period [yrs] inside the time bin
            if del_t <= 10**(8):
                integrated_photons += total_photons[itime]
            total_photons[itime] = total_photons[itime] / (del_t)
            
        else:
            del_t = (((10**(5.95 + 0.1 *(itime)))*(10**(0.1) - 1))) #calculating the time period [yrs] inside the time bin
            if del_t <= 10**(8):
                integrated_photons += total_photons[itime]
            total_photons[itime] = total_photons[itime] / (del_t)
            
    
    from scipy import interpolate
#     print(len(total_photons), len(time))
    time = np.logspace(6, 11, 51)
    intp_N_ion = interpolate.interp1d(time, total_photons, fill_value='extrapolate')
    t_new = np.arange(10**6, 10**11, 1e6)
    N_ion = np.array([intp_N_ion(t) for t in t_new])
    
    return [total_photons, delaytime, integrated_photons, N_ion, t_new]
```

```python
def count_ionising_photons(directory, file, index, how_many):
    df = pd.read_csv(directory + file, delim_whitespace=True,header = None)

    T = []
    for w in np.array_split(df, 3):
        T.append(w)
            
    total_photons = [ [0] for i in range(51)] # np.array([0]*51)        
    
    if how_many != 3: # if index = 3 then sum all.
        for j in range(51):
            total_photons[j] += T[index].iloc[j,0]
    else:
        for i in range(3):
            for j in range(51):
                total_photons[j] += T[i].iloc[j,0]
            
    # Binning them
    # integrated photons hold the total number of photons till 100 Mys.
    integrated_photons = 0
    delaytime = []
    for itime in range(51):
        delaytime.append(6 + 0.1*itime)
        
        if itime == 0:
            del_t = 10**(6.05) #calculating the time period [yrs] inside the time bin
            if del_t <= 10**(8):
                integrated_photons += total_photons[itime]
            total_photons[itime] = total_photons[itime] / (del_t*365.25*24*3600)
            
        else:
            del_t = (((10**(5.95 + 0.1 *(itime)))*(10**(0.1) - 1))) #calculating the time period [yrs] inside the time bin
            if del_t <= 10**(8):
                integrated_photons += total_photons[itime]
            total_photons[itime] = total_photons[itime] / (del_t*365.25*24*3600)
            
    
    return [total_photons, delaytime, integrated_photons]
```

```python
def GRB_count(directory, GRB_files):
    GRB_count_Z = [0]*len(GRB_files)
    typeIbc_count_Z = [0]*len(GRB_files)

    for i in range(len(GRB_files)):
        df_GRB = pd.read_csv(directory + GRB_files[i], header = None, delim_whitespace=True)
        G = []
        for w in np.array_split(df_GRB, 2): # CHE
            temp = []
            for x in np.array_split(w, 2): # GRB.
                temp.append(np.array_split(x, 2)) # typeIb/c
            G.append(temp)

        typeIbc = [0]*2
        GRB     = [0]*2
        for a in range(2):
            for b in range(2):
                typeIbc[a] += G[a][b][1].sum().sum() 
                for M in range(0,40):
                    M_min = math.floor(math.log10(df_tab.loc[i, 'M_min'])*10) + 9
                    M_max = math.ceil(math.log10(df_tab.loc[i, 'M_max'])*10) + 9
                    if b == 1 and M >= M_min and M <= M_max:
                        GRB[a] += sum(G[a][b][1].iloc[0:40, M]) # The G[a][b][0] wont count as typeIbc = 1 for GRB = 1

        GRB_count_Z[i] = GRB[1]  # as for a = 0 means no CHE so no GRBs
        typeIbc_count_Z[i] = typeIbc
        
    return GRB_count_Z  # This is an array whose first component is zero 
```
