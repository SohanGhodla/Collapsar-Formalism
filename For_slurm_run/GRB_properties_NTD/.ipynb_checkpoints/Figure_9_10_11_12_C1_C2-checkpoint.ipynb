# Calculationg the NS period

## Conditions
1. NS bariyonic mass - 1.6 Msun
2. A successful explosion
3. Assume moment of interia is fixed
4. J is conserved during core-collapse except the J taken away by neutrinos
    
From the above we get:
    $$P_{\rm NS} = 2\pi I_{\rm NS} / J_{\rm core}$$
    
where $I_{\rm NS} = 1.5 \times 10^{45}$ gcm$^{2}$

```python
dir_path = '/nesi/nobackup/uoa03218/Helium_stars_Dutch_variable/Makemodels_He_MS/'
```

```python
M_NS = 1.6
I_NS = 1.5*1e45
```

```python
mass =  np.arange(15,300,5.)
# mass = np.arange(15,300,5.)

Datafile = open('./Run_data_NS_period.txt' , 'w')
Datafile.write(f'Period_NS\t\t Z\t\t Mexp\t\t M_ZAMS\t\t E_rot\n')
for metal in range(len(z)):#len(metal)):
    for m in (mass):
        file_path = f'{dir_path}{z[metal]}/{str(m)}/LOGS/'
        if os.path.exists(file_path):
            df_p_index = pd.read_csv(file_path + '/profiles.index', delim_whitespace=True)

            #recording the last profile number
            last_profile = df_p_index.loc[len(df_p_index)-1,'lines']
            df_profile = pd.read_csv(file_path + 'profile' + str(last_profile) + '.data', delim_whitespace=True, header  = 4) 
            # Mexp_list[z].append(df_profile.loc[0,'mass'])

            # Checking for pair instability
            df_history = pd.read_csv(file_path + 'history.data', delim_whitespace=True, header  = 4)
            if df_history.loc[len(df_history)-1,'He_core'] >= 65. and  df_history.loc[len(df_history)-1,'He_core'] < 135.:
                continue  #then ignore the following

            fname = 'profile' + str(last_profile) 
            
            for i in range(len(df_profile)):
                if df_profile.loc[len(df_profile) - i -1, 'mass'] > M_NS:
                    Jcore = 10**df_profile.loc[len(df_profile) - i -1, 'log_J_inside']
                    break
                
            Period_NS = 2 * np.pi * I_NS / (0.57*Jcore)
            E_rot = 0.5 * I_NS * (2*np.pi/Period_NS)**2
            
            Period_NS = '%.3e' % Period_NS
            E_rot = '%.3e' % E_rot
            
            Mexp = '%.3e' % df_profile.loc[0,'mass']
#             print(f'{Period_NS}\t{metal[z]}\t{Mexp}\t{m}\t{E_rot}')
            Datafile.write(f'{Period_NS}\t\t{z[metal]}\t\t{Mexp}\t\t{m}\t\t{E_rot}\n')
Datafile.close()
```

## Calculating B_phi (flux conserved)

```python
z = ['zem5','zem4', '3zem4', '6zem4', 'z001', 'z002','z003', 'z004', 'z005']#, 'z006', 'z007', 'z008']
# mass =  [20, 21, 22, 23, 24, 25, 35, 40, 50, 60, 70, 80, 90, 100, 120, 150, 200, 250 , 300]
Z = [1e-5, 1e-4, 3e-4, 6e-4, 1e-3, 2e-3, 3e-3, 4e-3, 5e-3]
colors = plt.cm.Spectral(np.linspace(0,1,len(z)))
mass = np.arange(15,300,5.)
```

```python
file_path = f'{dir_path}{z[0]}/{str(15.)}/LOGS/'

df_p_index = pd.read_csv(file_path + 'profiles.index', delim_whitespace=True, low_memory=False)
#recording the last profile number
last_profile = df_p_index.loc[len(df_p_index)-1,'lines']
df_profile = pd.read_csv(file_path + 'profile' + str(last_profile) + '.data', delim_whitespace=True, header  = 4) 

df_profile.head()
```

```python
G = 6.667*1e-11 * 1e3  # to conver to cm3
Rsun = 696340 * 1e5 # cm
Msun = 1.98*1e30 * 1e3 # in gms
c = 3e8 *1e2  
Datafile = open('./B_radial_data.txt', 'w')
Datafile.write('B_radial\t\t Mfinal\t\t file\n')
for metal in range(len(z)):
    for m in (mass):
        file_path = f'{dir_path}{z[metal]}/{str(m)}/LOGS/'
        if os.path.exists(file_path):
            flux = 0
             # Checking for pair instability
            df_history = pd.read_csv(file_path + 'history.data', delim_whitespace=True, header  = 4)
            if df_history.loc[len(df_history)-1,'He_core'] >= 65. and  df_history.loc[len(df_history)-1,'He_core'] < 135.:
                continue  #then ignore the following


            df_p_index = pd.read_csv(file_path + 'profiles.index', delim_whitespace=True)
            #recording the last profile number
            last_profile = df_p_index.loc[len(df_p_index)-1,'lines']
            df_profile = pd.read_csv(file_path + 'profile' + str(last_profile) + '.data', delim_whitespace=True, header  = 4) 
            
            Mfinal = df_profile.loc[0,'mass']
            
            for i in range(len(df_profile)):
                temp = len(df_profile) - 1 - i
                M = df_profile.loc[temp, 'mass']  # Msun
                dr    = df_profile.loc[temp, 'dr']  # already in cm
                dm    = df_profile.loc[temp, 'dm']  # already in g
                r     = df_profile.loc[temp, 'radius'] * Rsun
                B_phi = 10**df_profile.loc[temp, 'dynamo_log_B_phi'] # Gauss
#                 B_phi = 10**df_profile.loc[temp, 'dynamo_log_B_r'] # Gauss
                
                
                if M <= 1.6:
                    flux += 2 * np.pi * B_phi * r * dr  # Gauss cm^2
#                     flux += B_phi * dm
                
            B_phi_NS = flux / (np.pi * (12e5)**2)  # 12 Km
#                 B_phi_NS = flux / (1.6 * (2*1e33))  # 12 Km
#             print('%.3e\t%.3e\t%s' %(B_phi_NS, Mfinal, f'{metal[z]}/{str(m)}'))
            Datafile.write('%.3e\t%.3e\t%s\n' %(B_phi_NS, Mfinal, f'{z[metal]}/{str(m)}'))
Datafile.close()             
```

```python
z = ['zem5', 'zem4',  '3zem4', '6zem4', 'z001', 'z002', 'z003', 'z004', 'z005']
Z = [1e-5, 1e-4, 3e-4, 6e-4, 1e-3, 2e-3, 3e-3, 4e-3, 5e-3]
colors = plt.cm.Spectral(np.linspace(0,1,len(z)))
```

```python
dirs = os.listdir(dir_path)
surf_avg_v_rot_at_TAMS = [[] for i in range(len(z))]
initial_rot = [[] for i in range(len(z))]
initial_surf_avg_v_rot = [[] for i in range(len(z))] 
omega_div_omega_crit_TAMS = [[] for i in range(len(z))] 
MZAMS = [[] for i in range(len(z))] 
M_final = [[] for i in range(len(z))] 
J_final = [[] for i in range(len(z))] 
c_12_final = [[] for i in range(len(z))]
he4_final = [[] for i in range(len(z))] 
mass_PPISNe = [[] for i in range(len(z))] 
he4_surf_final = [[] for i in range(len(z))] 
total_mass_he4 = [[] for i in range(len(z))] 

for metal in range(len(z)):
#     print(z[metal])
    masses = os.listdir(dir_path + z[metal])
    for m in range(len(masses)):
        directory = dir_path + z[metal] + '/' + masses[m] + '/LOGS' + '/'
        if os.path.isdir(directory):
            data = pd.read_csv(directory + 'history.data', header  = 4, delim_whitespace=True, low_memory=False)
            
            if data.loc[len(data) -1, 'He_core'] >= 65 and data.loc[len(data) -1, 'He_core'] < 133:
                mass_PPISNe[metal].append(data.loc[0, 'Mass'])
                continue
            else:
                mass_PPISNe[metal].append(0)

            initial_rot[metal].append(data.loc[0,'surf_avg_omega']) 
            initial_surf_avg_v_rot[metal].append(data.loc[0, 'surf_avg_v_rot'])

            omega_div_omega_crit_TAMS[metal].append(data.loc[len(data) -1,'surf_avg_omega_div_omega_crit']) #At CHE point
            MZAMS[metal].append(data.loc[0, 'Mass'])
            M_final[metal].append(data.loc[len(data) -1, 'Mass'])
            J_final[metal].append(data.loc[len(data) -1, 'log_total_angular_momentum'])
            c_12_final[metal].append(data.loc[len(data) -1, 'center_c12'])
            he4_final[metal].append(data.loc[len(data) -1, 'center_he4'])
            he4_surf_final[metal].append(data.loc[len(data) -1, 'surface_he4'])
            total_mass_he4[metal].append(data.loc[len(data) -1, 'total_mass_he4'])        
```

```python
plt.style.use('mesa.mplstyle13')
plt.figure(figsize = (4,5))
import matplotlib.gridspec as gridspec
gs = gridspec.GridSpec(3,1, hspace = 0.06, wspace = 0)

ax1 = plt.subplot(gs[0])
for i in range(len(Z)):
    plt.plot(MZAMS[i], he4_surf_final[i], marker = 's', markersize = 2.4,  linestyle = '', color = colors[i])

ax1.legend(Z, ncol = 3, bbox_to_anchor = (0.06,1.))
ax1.set_xlim(10,145)
ax1.set_ylabel('He$_{\\rm surf}$')
ax1.xaxis.set_ticklabels([])
ax1.grid(linestyle = '-.', alpha = 0.3)


ax2 = plt.subplot(gs[1])
for i in range(len(Z)):
    plt.plot(MZAMS[i], c_12_final[i], marker = 's', markersize = 2.4,  linestyle = '', color = colors[i])
# ax2.set_xlabel('M$_{\\rm ZAMS}$ [M$_{\\odot}$]')
ax2.xaxis.set_ticks_position('bottom')
ax2.set_ylabel('C$_{\\rm cntr}$')
ax2.set_xlim(10,145)
ax2.xaxis.set_ticklabels([])
ax2.grid(linestyle = '-.', alpha = 0.3)

ax3 = plt.subplot(gs[2])
for i in range(len(Z)):
    plt.plot(MZAMS[i], total_mass_he4[i], marker = 's', markersize = 2.4,  linestyle = '', color = colors[i])
ax3.set_xlabel('$M_{\\rm ZAMS}$ [M$_{\\odot}$]')
ax3.set_ylabel('$M_{\\rm He, total}$')
ax3.set_ylim(0.1,50)
ax3.xaxis.set_ticks_position('bottom')
ax3.set_xlim(10,120)
ax3.set_yscale('log')
ax3.grid(linestyle = '-.', alpha = 0.3)

plt.savefig('./Pics/He_surf_and_c_cntr_last_profile.jpg')
```
